```python
import pandas as pd
```

```python
df = pd.read_csv('PromoterSet.txt', sep='\t', header=None, comment='#')
df.columns = ["promoter_id", "name", "strand", "genome_pos", "sigma", "seq", "evidence", "conf_level"]
```

```python
mask1 = df['name'].str.contains('TSS')
df_TSS = df[mask1]
```

```python
df = df[~df['name'].astype(str).str.contains('TSS')]
```

**Questions Posed:**
- How many _unique_ genes have a promoter listed? Many genes have multiple promoters listed.
- How many unique genes have a promoter with STRONG confidence?

```python
df.shape
```

```python
mask = (df['conf_level'] == 'Strong')
```

```python
df_strong = df[mask]
df_strong.shape
```

Based on this cursory analysis, it looks like ~1700 genes have _strong_ confidence level for their promoter position. Let's look into this dataset a bit more deeply, to see if all of the genes are unique, and which type of evidence is frequently used to make these claims.

```python
df_strong.sample(20)
```

Hmm..it looks like the gene "yiaAp" has no sequence listed! That sucks. Let's remove any values for which NaN is present in the sequence column.

```python
df_strong_filter = df_strong.dropna(axis=0, how='any')
df_strong_filter.shape
```
