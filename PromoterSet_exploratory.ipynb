**Questions Posed:**
- How many _unique_ genes have a promoter listed? Many genes have multiple promoters listed.
- How many unique genes have a promoter with STRONG confidence?

```python
df.shape
```

```python
mask = (df['conf_level'] == 'Strong')
```

```python
df_strong = df[mask]
df_strong.shape
```

Based on this cursory analysis, it looks like ~1700 genes have _strong_ confidence level for their promoter position. Let's look into this dataset a bit more deeply, to see if all of the genes are unique, and which type of evidence is frequently used to make these claims.

The reviewer also asked us to consider how likely it is that each operon is governed by a _single_ promoter. To get a "ballpark" estimate, let's look at the genome positions, and filter out those that are within 200bp of one another...

To do this, I will actually first need to make two DataFrames...one for the promoters that are "FORWARD" and another for the promoters that are "REVERSE", since forward and reverse promoters frequently occur next to one another, but regulate very different operons.

```python
df_fwd = df_strong[df_strong['strand'] == 'forward']
df_fwd.shape
```

```python
df_rev = df_strong[df_strong['strand'] == 'reverse']
df_rev.shape
```

```python
df_fwd_sort = df_fwd['genome_pos'].sort_values()
```

```python
df_rev_sort = df_rev['genome_pos'].sort_values()
```

```python
df_fwd_diff = df_fwd_sort.diff()
```

```python
df_rev_diff = df_rev_sort.diff()
```

```python
len(df_fwd_diff)
```

```python
len(df_rev_diff)
```

Now let's apply an arbitrary filter to our sorted Series that pulls out only those values that are more than 500bp apart, as we can assume these are for unique genes.

```python
filter = df_fwd_diff > 500
```

```python
df_fwd_diff[filter]
```

```python
filter2 = df_rev_diff > 500
df_rev_diff[filter2]
```
