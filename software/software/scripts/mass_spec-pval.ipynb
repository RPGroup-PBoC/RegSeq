```python
#import basic modules
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
#load in all mass spec proteinGroup files. These contain the normalized heavy to light ratios for all
#identified proteins.
allnames = glob.glob('massspec/protein*.txt')
```

```python
#we can also load in the proteinGroup files for only the targets we will be summarizing.
all_filtered = glob.glob('massspec/filtered/*')
```

```python
#We will compile all ratios from all mass spec runs. We will organize these by the protein name. 
fulldf = pd.DataFrame()
for i,n in enumerate(allnames):
    tempdf = pd.io.parsers.read_csv(n,sep='\t')
    columns = tempdf.columns
    #We only want those columns with the normalized ratios of Heavy to light lysate
    good_cols = [x for x in columns if 'normalized ' in x if 'Heavy' not in x]
    #now we loop through each of those good columns and add them into a final output df with all appropriate columns
    for i, col in enumerate(good_cols):
        tempdf2 = tempdf[['Protein names','Gene names'] + [col]].copy()
    
        tempdf2 = tempdf2.dropna()
        #append column to the output 'fulldf'
        fulldf = pd.concat([fulldf,tempdf2],axis=0)
    
    print(fulldf)
```

```python
#for any missing ratios we will fill the value with a 0. 
fulldf2 = fulldf.fillna(value=0)
```

```python
#this will set the Protein names as the index value.
fulldf3 = fulldf2.groupby('Protein names').sum()
```

```python
#we will format an output dataframe that contains a mean value and variance the most highly enriched protein
#and for all background proteins.
out_df = pd.DataFrame(columns=['mean','var'])
out_background_df = pd.DataFrame(columns=['mean','var'])
#create a dataframe for pvals
out_pval = pd.DataFrame(columns=['pval'])
#format the output look of each dataframe.
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.float_format', '{:10,.9f}'.format)
```

```python
#We will loop through all enriched proteins displayed in the figures in the Reg-Seq paper.
for z,filtered_name in enumerate(all_filtered):
    #load in file with proteins and enrichments
    indf = pd.io.parsers.read_csv(filtered_name)
    #get the correct column name that contains the heavy to light ratio.
    indf_ratio_col = indf.columns[-1]
    row = indf.loc[0,:]
    name = row['Protein names']

    #get every example of a protein enrichment ratio from the protein of interest, we will use this
    #to calculate that typical uncertainty in enrichment ratios we get when measuring this protein.
    q = fulldf3.loc[name]
    #get the enrichment ratio of the protein in this experiment. 
    theratio = float(row[indf_ratio_col])
    
    #now we will look at the background measurements, i.e. all other proteins but the most enriched.
    #first we get rows where this is the case
    goodrows = (q.index != indf_ratio_col)
    
    #filter dataframe to only have these entries and also drop any with an enrichment ratio of zero.
    q = q[goodrows]
    q = q[q > 1e-8]
    
    #now we calculate the p_value that the most enriched protein is actually enriched compared to background.
    
    #we will calculate this value on a log scale.
    
    #calculate mean and variance on a log scale for the protein of interest.
    tempmean = np.log(theratio)
    tempvar = np.var(np.log(q))
    
    out_df.loc[indf_ratio_col,:] = [tempmean,tempvar]
    
    #calculate the same measures for the background proteins.
    background_df = indf.loc[1:,:]
    background_tempmean = np.mean(np.log(background_df.loc[:,indf_ratio_col]))
    background_tempvar = np.var(np.log(background_df.loc[:,indf_ratio_col]))
    
    
    out_background_df.loc[str(indf_ratio_col),:] = [float(background_tempmean),float(background_tempvar)]
    
    #calculae the p-value
    statistic,pval = scipy.stats.ttest_ind_from_stats(tempmean,tempvar,len(q),background_tempmean,background_tempvar,len(background_df.index))
    
    #write results to file.
    with open('test_pval','a') as f:
        f.write(indf_ratio_col + ',' + str(pval) + '\n')
    out_pval.loc[indf_ratio_col,:] = [pval]
```
