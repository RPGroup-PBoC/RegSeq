```python
import pandas as pd
import glob
import numpy as np
```

```python
names = glob.glob('*smooth*')
```

```python
names
```

```python
def sliding_window(y,windowsize=3):
    out_vec = np.zeros_like(y)
    for i in range(len(y)-windowsize):
        out_vec[i] = np.sum(y[i:i+windowsize])/windowsize
    return out_vec
```

```python
infodf = pd.DataFrame()
infodf['pos'] = range(1,158)
#infodf['pos'] = range(160)
```

```python
possible_growths = ['arabinose','hdfR','M9','SS','LB','Anaero','glpR','heat']
```

```python
for i,name in enumerate(names):
    tempdf = pd.io.parsers.read_csv(name,delim_whitespace=True)
    for x in possible_growths:
        if x in name:
            gene = name.split(x)[0]
            growth = x
    if growth == 'arabinose':
        tempdf['info'] = sliding_window(tempdf['info'])
    infodf[gene + '_' + growth + '_info'] = tempdf['info']
```
